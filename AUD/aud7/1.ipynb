```python
X = df.drop(['id','outcome'],axis=1)
Y = df['outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
cv.fit(X,Y)
```

```python
cv.best_score_
```

```python
model = cv.best_estimator_
```

```python
y_pred = model.predict(X_test)
```
