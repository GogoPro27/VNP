# Local Elections 2021

For each municipality, scrape:
- Municipality name ("Општина")
- Political party name ("Учесници во изборите")
- Number of votes received by the party ("Гласови")
- Percentage of votes received by the party ("Процент")
- Registered voters ("Регистрирани гласачи")

- Number of invalid votes ("Неважечки гласови")

```python
import time
time.sleep(1)

mydic = {}

for i in range(0,7):
    redovi = browser.find_elements(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(7) > div > div.hidden.md\\:block > div > div > div > div")
    j = 0
    for j in range(2, len(redovi) - 1):
        link = redovi[j].find_element(By.CSS_SELECTOR, "div:nth-child(1) > a")
        opstina = link.text
        mydic[opstina] = link.get_attribute("href")
    
    button = browser.find_element(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(7) > div > div.hidden.md\\:block > div > div > div > div.grid.grid-cols-12.mt-10 > div > div > button:nth-child(4)")
    button.click()

opstini = []

for link in mydic.values():
    browser.get(link)
    time.sleep(1)
    #     Municipality name ("Општина")
    name = browser.find_element(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div.grid.grid-cols-12.gap-5.pt-4 > div > p").text
    
    wrapper_partii_redovi = browser.find_elements(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(5) > div > div.grid.grid-cols-1 > div > div > div")
    partii=[]
    for a in range(2,len(wrapper_partii_redovi)):
        ime = wrapper_partii_redovi[a].find_element(By.CSS_SELECTOR,"div.col-span-9.md\\:col-span-8 > div > div > div > div:nth-child(1)").text
        ime = ime[:len(ime)//2]
        glasovi = wrapper_partii_redovi[a].find_element(By.CSS_SELECTOR,"div.col-span-9.md\\:col-span-8 > div > div > div > div:nth-child(2)").text.split(" ")[0]
        procent = wrapper_partii_redovi[a].find_element(By.CSS_SELECTOR,"div.col-span-3.md\\:col-span-4 > div > div.col-span-3.md\\:col-span-1 > p").text
        partii.append({"IME":ime,"GLASOVI":glasovi,"PROCENT":procent})
    
    # - Political party name ("Учесници во изборите")
    # - Number of votes received by the party ("Гласови")
    # - Percentage of votes received by the party ("Процент")
    
    # - Registered voters ("Регистрирани гласачи")
    reg_voters = browser.find_element(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(6) > div > div.grid.grid-cols-1 > div > div > div:nth-child(1) > div > div.col-span-1.flex.justify-end > p").text
    # - Number of valid votes ("Важечки гласови")
    acc_voters = browser.find_element(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(6) > div > div.grid.grid-cols-1 > div > div > div:nth-child(3) > div > div.col-span-1.flex.justify-end > p").text
    # - Number of invalid votes ("Неважечки гласови")
    not_acc_voters = browser.find_element(By.CSS_SELECTOR,"#root > div > main > div:nth-child(1) > div.lg\\:container.mx-auto.px-3.xl\\:px-40 > div:nth-child(6) > div > div.grid.grid-cols-1 > div > div > div:nth-child(4) > div > div.col-span-1.flex.justify-end > p").text
    print(name,partii,reg_voters,acc_voters,not_acc_voters)
    print()
    print()
    


```
